```python
from diagrams import Cluster, Diagram
from diagrams.aws.storage import S3
from diagrams.aws.database import Redshift
from diagrams.aws.compute import EC2
from diagrams.onprem.analytics import Spark

"""
This piece of code creates a diagram image which im going to use in the readme to visualize the project
"""


with Diagram("Sparkify Data Architecture", show=False) as med_diag:
    with Cluster("Redshift Database"):
        redshift_staging = Redshift("Staging")
        redshift_bi = Redshift("Target")
        
        
    with Cluster("Datalake"):
        S3_pagila = S3("S3")                            

    with Cluster("Queries"):
        query_s3_redshift = EC2("SQL")
        query_redshift_redshift = EC2("SQL")
    with Cluster("Analytics"):
        spark = Spark()

    S3_pagila >> query_s3_redshift >> redshift_staging >> query_redshift_redshift >> redshift_bi  >> spark

    
med_diag
```
